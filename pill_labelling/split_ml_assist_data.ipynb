```python
import pandas as pd
```

```python
df = pd.read_csv("../data/splits/ml_assisted_1800.csv")
df.head(3)
```

```python
import os

# Shuffle the dataframe with random seed 42
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Create output directory
output_dir = "../data/splits/ml_labelling_batches"
os.makedirs(output_dir, exist_ok=True)

# Define custom batch sizes: 500, 500, 800, then remaining
batch_sizes = [500, 500, 800]
start_idx = 0
batch_num = 1

# Create the specified batches
for batch_size in batch_sizes:
    if start_idx >= len(df_shuffled):
        break

    end_idx = min(start_idx + batch_size, len(df_shuffled))
    batch_df = df_shuffled.iloc[start_idx:end_idx]

    filename = f"{output_dir}/ml_labelling_batch_{batch_num}.csv"
    batch_df.to_csv(filename, index=False)
    print(f"Saved {len(batch_df)} rows to {filename}")

    start_idx = end_idx
    batch_num += 1

# Handle any remaining samples (if total > 1800)
if start_idx < len(df_shuffled):
    batch_df = df_shuffled.iloc[start_idx:]
    filename = f"{output_dir}/ml_labelling_batch_{batch_num}.csv"
    batch_df.to_csv(filename, index=False)
    print(f"Saved {len(batch_df)} rows to {filename}")
    batch_num += 1

print(f"\nTotal: {batch_num - 1} batches created")
```
